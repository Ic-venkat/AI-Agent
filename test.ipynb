```python
from dotenv import load_dotenv, find_dotenv

load_dotenv(find_dotenv(), override=True)
```

```python
import os
from crewai import Agent, Task, Crew, Process
from langchain.agents import Tool
from crewai_tools import WebsiteSearchTool
from langchain_experimental.utilities import PythonREPL

from langchain_community.tools import DuckDuckGoSearchRun

search_tool = DuckDuckGoSearchRun()
website_search_tool = WebsiteSearchTool()
repl_tool = Tool(
    name="python_repl",
    description="A python shell. Use this to execute python commands. You can also ask input from a user using this.",
    func=PythonREPL().run
)
```

```python
scraper = Agent(
    role="Summarizer of websites",
    goal="Go to a website, scrape the content and provide the full content to the writer agent",
    backstory="""You work at a leading IT consultancy,
    your expertise lies in taking URLs and getting text-based content of them.""",
    verbose=True,
    allow_delegation=False,
    tools=[search_tool, website_search_tool]
)

writer = Agent(
    role="Tech content summarizer and writer",
    goal="""craft compelling content on tech advancements based on text provided to you""",
    backstory="""You are a renowned content creator working at a leading tech consultancy called Cloudxcel,
    known for you insightful and engaging articles.
    You transform complex articles into compelling narratives""",
    verbose=True,
    allow_delegation=True
)

task1 = Task(
    description="""Take a list of websites that contain AI content, read scrape 
    the content and pass it to the writer agent""",
    agent=scraper,
    expected_output="Full context of all articles provided to the scraper"
)

task2 = Task(
    description="""Using the text provided by the scraper, develop an engaging blog post
    that highlights the most significant AI advancements.
    Your post is informative yet accessible, catering the techsavvy audience""",
    agent=writer,
    expected_output="Full blog post from the context"
)

crew = Crew(
    agents=[scraper, writer],
    tasks=[task1, task2],
    verbose=2
)

result = crew.kickoff()
```

```python
print(result)
```
